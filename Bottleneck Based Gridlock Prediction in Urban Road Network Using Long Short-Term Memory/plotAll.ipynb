```python
import csv
import os
import pandas as pd
import math

percentage = ['1%', '5%', '10%', '15%', '20%', '25%', '30%', '35%', '40%', '45%', '50%']
resolution = ['1','5','10','15','20','25','30','35','40','45','50','55','60']


# def plotAllStandardErrorOfMean(): ######################## this function has been modified in statisticsForAll.py
#     import pandas as pd
#     import matplotlib.pyplot as plt

#     dirpath = os.getcwd()
#     path = dirpath + '/All_stdErrorOfMeanForMultipleRuns.csv'
#     if (os.path.exists(path)):
#         df = pd.read_csv(path)    
        
        
#         colors = ['#CD5C5C', '#808080', '#808000', '#0000FF', '#000080', '#BDB76B', '#3CB371', '#D8BFD8',
#                   '#5F9EA0', '#7B68EE', '#D8BFD8', '#7B68EE', '#800000', '#FFC0CB', '#808080',
#                   '#DAA520', '#BC8F8F', '#2F4F4F', '#008000', '']
# #         colors = ['#FFA07A', '#FF7F50', '#FFFFE0', '#98FB98', '#E0FFFF', '#B0E0E6', '#E6E6FA']
#         names = resolution
           
#         df.set_index('Samples', inplace = True) 
       

#         ax = df[[*names]].plot(kind='bar',figsize=(15, 10), legend=True, color=colors)
#         ax.set_xlabel('Samples' , fontsize=20, fontname ='Times New Roman')
#         ax.set_ylabel("Standard Error of Mean", fontsize=20, fontname ='Times New Roman')
#         ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'family' :'Times New Roman','size':20},title="Time Resoultion (s)")
# #         plt.legend(prop={'family' :'Times New Roman'})
#         plt.xticks(fontsize=20, fontname = 'Times New Roman',rotation=0)
#         plt.yticks(fontsize=20,fontname = 'Times New Roman')
#         #plt.rcParams.update({'font.family':'Times New Roman'})
#         plt.tight_layout()
#         plt.savefig(
#             dirpath + '/All_stdErrorOfMeanForMultipleRuns.png',
#             width=1800, height=500)



def plotDetectionRateWithBarPlot():
    import pandas as pd
    import matplotlib.pyplot as plt

    dirpath = os.getcwd()

    path = dirpath + '/detectionMetricForAllClassLabel_ForMultipleRuns.csv'
    if (os.path.exists(path)):
        all_df = pd.read_csv(path)
        #colors = ['#616A6B','#ABB2B9']
        colors = ['#616A6B']
        
        for label in range(0,5):
        
        
            #names = ["DR_PG", "DR_G"]
            names = ["DR_"+str(label+1)]
            #https://stackoverflow.com/questions/29498652/plot-bar-graph-from-pandas-dataframe
            df = all_df.set_index("Samples")

            ax = df[[*names]].plot(kind='bar',figsize=(15, 10), legend=True, color=colors, width =0.8)
            ax.set_xlabel("Samples", fontsize=20, fontname ='Times New Roman')
            ax.set_ylabel("Detection Rate (%)", fontsize=20, fontname ='Times New Roman')
            ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'family' :'Times New Roman','size':20})

            ################################################################################
            #https://robertmitchellv.com/blog-bar-chart-annotations-pandas-mpl.html
            # create a list to collect the plt.patches data
            totals = []

            # find the values and append to list
            for i in ax.patches:
                totals.append(i.get_height())

            # set individual bar lables using above list
            for i in ax.patches:
                # get_x pulls left or right; get_height pushes up or down
                ax.text(i.get_x() - .001, i.get_height() + .5, \
                        str(int(round((i.get_height())))) + '%', fontsize=10,
                        color='red')
            ################################################################################

            #plt.legend(prop={'family' :'Times New Roman'})
            plt.xticks(fontsize=20, fontname = 'Times New Roman',rotation=0)
            plt.yticks(fontsize=20,fontname = 'Times New Roman')
            #plt.rcParams.update({'font.family':'Times New Roman'})
            plt.tight_layout()
            plt.savefig(
                dirpath + '/detectionMetricForAllClassLabel_ForMultipleRuns'+str(label+1)+'.png',
                width=1800, height=500)

def plotFalseAlarmRateWithBarPlot():
    import pandas as pd
    import matplotlib.pyplot as plt

    dirpath = os.getcwd()
    
    path = dirpath + '/falseAlarmRateMetricForAllClassLabel_ForMultipleRuns.csv'
    if (os.path.exists(path)):
        all_df = pd.read_csv(path)
        #colors = ['#009E73', '#E69F00']
        colors = ['#009E73']
        for label in range(0,5):
            #names = ["FAR_PG","FAR_G"]
            names = ["FAR_"+str(label+1)]
            #https://stackoverflow.com/questions/29498652/plot-bar-graph-from-pandas-dataframe
            df = all_df.set_index("Samples")

            ax = df[[*names]].plot(kind='bar',figsize=(15, 10), legend=True, color=colors, width =0.8)
            ax.set_xlabel("Samples", fontsize=20, fontname ='Times New Roman')
            ax.set_ylabel("False Alarm Rate (%)", fontsize=20, fontname ='Times New Roman')
            ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'family' :'Times New Roman','size':20})

            ################################################################################
            #https://robertmitchellv.com/blog-bar-chart-annotations-pandas-mpl.html
            # create a list to collect the plt.patches data
            totals = []

            # find the values and append to list
            for i in ax.patches:
                totals.append(i.get_height())

            # set individual bar lables using above list
            for i in ax.patches:
                # get_x pulls left or right; get_height pushes up or down
                ax.text(i.get_x() - .001, i.get_height() + .5, \
                        str(int(round((i.get_height())))) + '%', fontsize=10,
                        color='red')
            ################################################################################

            #plt.legend(prop={'family' :'Times New Roman'})
            plt.xticks(fontsize=20, fontname = 'Times New Roman',rotation=0)
            plt.yticks(fontsize=20,fontname = 'Times New Roman')
            
            #plt.rcParams.update({'font.family':'Times New Roman'})
            plt.tight_layout()
            plt.savefig(
                dirpath + '/falseAlarmRateMetricForAllClassLabel_ForMultipleRun'+str(label+1)+'.png',
                width=1800, height=500)

            
            
def plotFalseAlarmRateWithBoxPlot():
    import pandas as pd
    import matplotlib.pyplot as plt

    dirpath = os.getcwd()
    
    for i in range(1,6):
    
        path = dirpath + '/falseAlarmRateForAllClassLabel_'+str(i)+'ForMultipleRuns.csv'
        if (os.path.exists(path)):
            all_df = pd.read_csv(path)           

            names =  ['1%','5%','10%','15%','20%','25%','30%']
            #https://stackoverflow.com/questions/29498652/plot-bar-graph-from-pandas-dataframe
            #df = all_df.set_index("Samples")
            props = dict(boxes="DarkGreen", whiskers="DarkOrange", medians="DarkBlue", caps="Gray")
            ax = all_df[[*names]].plot(kind='box',figsize=(15, 10),color=props)
            ax.set_xlabel("Samples", fontsize=20, fontname ='Times New Roman')
            ax.set_ylabel("False Alarm Rate (%)", fontsize=20, fontname ='Times New Roman')
            #ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), prop={'family' :'Times New Roman','size':20} , labels  =names )

            ################################################################################
            #https://robertmitchellv.com/blog-bar-chart-annotations-pandas-mpl.html
            # create a list to collect the plt.patches data
            totals = []

            # find the values and append to list
            for i in ax.patches:
                totals.append(i.get_height())

            # set individual bar lables using above list
            for i in ax.patches:
                # get_x pulls left or right; get_height pushes up or down
                ax.text(i.get_x() - .001, i.get_height() + .5, \
                        str(int(round((i.get_height())))) + '%', fontsize=10,
                        color='red')
            ################################################################################

            #plt.legend(prop={'family' :'Times New Roman'})
            plt.xticks(fontsize=20, fontname = 'Times New Roman',rotation=0)
            plt.yticks(fontsize=20,fontname = 'Times New Roman')
            #plt.title('False Alarm Rate For Gridlock Label '+ str(i),fontsize=20,fontname = 'Times New Roman')
            #plt.rcParams.update({'font.family':'Times New Roman'})
            plt.tight_layout()
            plt.savefig(
                dirpath + '/falsearamrateMetricForAllClassLabel_'+str(i)+'ForMultipleRuns_withBoxPlot.png',
                width=1800, height=500)
        
        
        
def plotDetectionRateWithBoxPlot():
    import pandas as pd
    import matplotlib.pyplot as plt

    dirpath = os.getcwd()
    
    for i in range(1, 6):
    
        path = dirpath + '/detectionMetricForAllClassLabel_'+str(i)+'ForMultipleRuns.csv'
        if (os.path.exists(path)):
            all_df = pd.read_csv(path)           

            names =  ['1%','5%','10%','15%','20%','25%','30%']
            #https://stackoverflow.com/questions/29498652/plot-bar-graph-from-pandas-dataframe
            #df = all_df.set_index("Samples")
            #whiskerprops = dict(linewidth=1.0, color='DarkGreen')
            props = dict(boxes="DarkGreen", whiskers="DarkOrange", medians="DarkBlue", caps="Gray")
            ax = all_df[[*names]].plot(kind='box',figsize=(15, 10), color=props)
            ax.set_xlabel("Samples", fontsize=20, fontname ='Times New Roman')
            ax.set_ylabel("Detection Rate (%)", fontsize=20, fontname ='Times New Roman')
            
            #ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), prop={'family' :'Times New Roman','size':20}  )

            ################################################################################
            #https://robertmitchellv.com/blog-bar-chart-annotations-pandas-mpl.html
            # create a list to collect the plt.patches data
            totals = []

            # find the values and append to list
            for i in ax.patches:
                totals.append(i.get_height())

            # set individual bar lables using above list
            for i in ax.patches:
                # get_x pulls left or right; get_height pushes up or down
                ax.text(i.get_x() - .001, i.get_height() + .5, \
                        str(int(round((i.get_height())))) + '%', fontsize=10,
                        color='red')
            ################################################################################

            #plt.legend(prop={'family' :'Times New Roman'})
            plt.xticks(fontsize=20, fontname = 'Times New Roman',rotation=0)
            plt.yticks(fontsize=20,fontname = 'Times New Roman')
            #plt.title('Detection Rate For Gridlock Label '+ str(i),fontsize=20,fontname = 'Times New Roman')
            #plt.rcParams.update({'font.family':'Times New Roman'})
            plt.tight_layout()
            plt.savefig(
                dirpath + '/detectionMetricForAllClassLabel_'+str(i)+'_ForMultipleRuns_withBoxPlot.png',
                width=1800, height=500)

            
            
def plotDetectionRateAll_InOneGraph():
    import pandas as pd
    import matplotlib.pyplot as plt

    dirpath = os.getcwd()

    path = dirpath + '/detectionMetricForAllClassLabel_ForMultipleRuns.csv'
    if (os.path.exists(path)):
        all_df = pd.read_csv(path)
        #colors = ['#616A6B','#ABB2B9']
        colors = ['#CD5C5C', '#808080', '#808000', '#BDB76B', '#3CB371']
        
#         for label in range(0,5):
        
        
        #names = ["DR_PG", "DR_G"]
        names = ["DR_1","DR_2","DR_3","DR_4","DR_5"]
        #https://stackoverflow.com/questions/29498652/plot-bar-graph-from-pandas-dataframe
        df = all_df.set_index("Samples")

        ax = df[[*names]].plot(kind='bar',figsize=(15, 10), legend=True, color=colors, width =0.8)
        ax.set_xlabel("Samples", fontsize=20, fontname ='Times New Roman')
        ax.set_ylabel("Detection Rate (%)", fontsize=20, fontname ='Times New Roman')
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'family' :'Times New Roman','size':20})

        ################################################################################
        #https://robertmitchellv.com/blog-bar-chart-annotations-pandas-mpl.html
        # create a list to collect the plt.patches data
        totals = []

        # find the values and append to list
        for i in ax.patches:
            totals.append(i.get_height())

        # set individual bar lables using above list
        for i in ax.patches:
            # get_x pulls left or right; get_height pushes up or down
            ax.text(i.get_x() - .001, i.get_height() + .5, \
                    str(int(round((i.get_height())))) + '%', fontsize=10,
                    color='red')
        ################################################################################

        #plt.legend(prop={'family' :'Times New Roman'})
        plt.xticks(fontsize=20, fontname = 'Times New Roman',rotation=0)
        plt.yticks(fontsize=20,fontname = 'Times New Roman')
        #plt.rcParams.update({'font.family':'Times New Roman'})
        plt.tight_layout()
        plt.savefig(
            dirpath + '/detectionMetricForAllClassLabel_ForMultipleRuns.png', width=1800, height=500)
  
def plotFalseAlarmRateAll_InOneGraph():
    import pandas as pd
    import matplotlib.pyplot as plt

    dirpath = os.getcwd()

    path = dirpath + '/falseAlarmRateMetricForAllClassLabel_ForMultipleRuns.csv'
    if (os.path.exists(path)):
        all_df = pd.read_csv(path)
        #colors = ['#616A6B','#ABB2B9']
        colors = ['#CD5C5C', '#808080', '#808000', '#BDB76B', '#3CB371']
        
#         for label in range(0,5):
        
        
        #names = ["DR_PG", "DR_G"]
        names = ["FAR_1","FAR_2","FAR_3","FAR_4","FAR_5"]
        #https://stackoverflow.com/questions/29498652/plot-bar-graph-from-pandas-dataframe
        df = all_df.set_index("Samples")

        ax = df[[*names]].plot(kind='bar',figsize=(15, 10), legend=True, color=colors, width =0.8)
        ax.set_xlabel("Samples", fontsize=20, fontname ='Times New Roman')
        ax.set_ylabel("False Alarm Rate (%)", fontsize=20, fontname ='Times New Roman')
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'family' :'Times New Roman','size':20})

        ################################################################################
        #https://robertmitchellv.com/blog-bar-chart-annotations-pandas-mpl.html
        # create a list to collect the plt.patches data
        totals = []

        # find the values and append to list
        for i in ax.patches:
            totals.append(i.get_height())

        # set individual bar lables using above list
        for i in ax.patches:
            # get_x pulls left or right; get_height pushes up or down
            ax.text(i.get_x() - .001, i.get_height() + .5, \
                    str(int(round((i.get_height())))) + '%', fontsize=10,
                    color='red')
        ################################################################################

        #plt.legend(prop={'family' :'Times New Roman'})
        plt.xticks(fontsize=20, fontname = 'Times New Roman',rotation=0)
        plt.yticks(fontsize=20,fontname = 'Times New Roman')
        #plt.rcParams.update({'font.family':'Times New Roman'})
        plt.tight_layout()
        plt.savefig(
            dirpath + '/falseAlarmRateMetricForAllClassLabel_ForMultipleRuns.png', width=1800, height=500)



#main functin
#========================================================================================
def main():

#     plotDetectionRateWithBarPlot()
#     plotFalseAlarmRateWithBarPlot()    
#     plotFalseAlarmRateWithBoxPlot()
#     plotDetectionRateWithBoxPlot()
    plotDetectionRateAll_InOneGraph()
    plotFalseAlarmRateAll_InOneGraph()
   

if __name__=="__main__":
    main()
```
