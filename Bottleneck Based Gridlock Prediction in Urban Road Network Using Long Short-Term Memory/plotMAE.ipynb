```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import pandas as pd


dirpath = os.getcwd()
percent = ['5%','10%','15%','20%','25%','30%','35%','40%','45%','50%']
path  = dirpath + '/LSTMExperimentResults_AfterDefense/MAE_Confidence_Interval (epoch= 10, batch = 50 , neurons = 10) - without-1percent.csv'
names =['Samples','Number of Observations','MAEValues','Mean','Standard Errors','Upper  Bound','Lower Bound','Execution Time']
plt.figure(figsize=(8,5))
if (os.path.isfile(path)):
    df = pd.read_csv(path, names=names, header=None, skiprows=1)    
    
    select_df = df.loc[df['Number of Observations']==  1]
    one_min = select_df['Mean'].tolist()
    select_df = df.loc[df['Number of Observations']==  5]
    five_min = select_df['Mean'].tolist()
    select_df = df.loc[df['Number of Observations']==  10]
    ten_min = select_df['Mean'].tolist()
    select_df = df.loc[df['Number of Observations']==  15]
    fifteen_min = select_df['Mean'].tolist()
    select_df = df.loc[df['Number of Observations']==  30]
    thirty_min = select_df['Mean'].tolist()
    select_df = df.loc[df['Number of Observations']==  60]
    sixty_min = select_df['Mean'].tolist()
    
    plt.plot(percent, one_min, '-rD',label = "1 min")
    plt.plot(percent, five_min, '-kD',label = "5 min")
    plt.plot(percent, ten_min, '-bD',label = "10 min")
    plt.plot(percent, fifteen_min, '-cD',label = "15 min") 
    plt.plot(percent, thirty_min, '-mD',label = "30 min")
    plt.plot(percent, sixty_min, '-gD',label = "60 min")
    plt.legend(loc='center left', bbox_to_anchor= (1,0.5), title ="time-lagged observation period",prop={'family' :'Times New Roman','size':20})
    
    plt.xlabel ('Samples',fontsize=20, fontname ='Times New Roman')
    plt.ylabel('MAE',fontsize=20, fontname ='Times New Roman')
    plt.tight_layout()
    plt.savefig(dirpath+'/LSTMExperimentResults_AfterDefense/MAE_Confidence_Interval (epoch= 10, batch = 50 , neurons = 10) - without-1percent.png')  
```
